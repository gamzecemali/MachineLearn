```python
import numpy as np
import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.arange(0, 1, 0.1)
y = np.sin(x)

plt.plot(x, y)
plt.title("Sin√ºs ")
plt.xlabel("X ekseni")
plt.ylabel("Y ekseni")
plt.show()
```

```python
x = np.random.rand(10)
y = np.random.rand(10)
colors = np.random.rand(10)
sizes = np.random.randint(10, 100, 10)

plt.scatter(x, y, c=colors, s=sizes)
plt.title("Point Chart")
plt.xlabel("X-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
data = np.random.randn(200)

plt.hist(data, bins=10)
plt.title("hist")
plt.xlabel("deger")
plt.ylabel("frekans")
plt.show()
```

```python
x = ['A', 'B', 'C', 'D', 'E']
y = np.random.randint(3, 10, 5)

plt.bar(x, y)
plt.title("Column Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
sizes = [85, 30, 20, 60, 15, 79]

plt.pie(sizes)
plt.title("Pie Chart")
plt.show()
```

```python
x = np.arange(-90, 3, 0.10)
y = np.arange(-90, 15, 0.10)
x, y = np.meshgrid(x, y)

r = np.sqrt(x**2 + y**2)
z = np.sin(r)

fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z)
ax.set_title("3D Graphics")
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.set_zlabel("Z axis")

plt.show()
```
