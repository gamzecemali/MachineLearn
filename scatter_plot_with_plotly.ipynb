```python
timesData = pd.read_csv("data/timesData.csv")
```

```python
timesData.head()
```

```python
df2014 = timesData[timesData.year == 2014].iloc[:100,:] 
df2015 = timesData[timesData.year == 2015].iloc[:100,:] 
df2016 = timesData[timesData.year == 2016].iloc[:100,:] 
```

```python

trace1 =go.Scatter( 
                    x = df2014.world_rank, 
                    y = df2014.citations,  
                    mode = "markers",      
                    name = "2014",         
                    marker = dict(color = 'rgba(255, 128, 255, 0.8)'),
                    text= df2014.university_name)  

# creating trace2
trace2 =go.Scatter(
                    x = df2015.world_rank,
                    y = df2015.citations,  
                    mode = "markers",      
                    name = "2015",        
                    marker = dict(color = 'rgba(255, 128, 2, 0.8)'),  
                    text= df2015.university_name)  


trace3 =go.Scatter(
                    x = df2016.world_rank, 
                    y = df2016.citations,  
                    mode = "markers",      
                    name = "2016",         
                    marker = dict(color = 'rgba(0, 255, 200, 0.8)'), 
                    text= df2016.university_name) 

data = [trace1, trace2, trace3]  

layout = dict(title = 'Citation vs world rank of top 100 universities with 2014, 2015 and 2016 years',
              xaxis= dict(title= 'World Rank',ticklen= 5,zeroline= False),
              yaxis= dict(title= 'Citation',ticklen= 5,zeroline= False)  
             )
fig = dict(data = data, layout = layout)  

iplot(fig) 
```

```python
# creating trace1 -> 2014,2015 ve 2016 olmak zere  farkl data olduundan 3 farkl trace yaratlacaktr.
trace1 =go.Scatter( #scatter tr seilir.
                    x = df2014.world_rank, #x eksenine 2014 ylndaki ilk 100 niversite koyulur.
                    y = df2014.total_score,  #y eksenine 2014 ylndaki alnt deerleri koyulur.
                    mode = "markers",      #marker modu seilir, yani nokta ekli
                    name = "2014",         #trace'in ad verilir.
                    marker = dict(color = 'rgba(255, 128, 255, 0.8)'), #renk ve saydamlk verilir.
                    text= df2014.university_name)  #bir deerin zerine gelindiinde niversite ad gzkr.

# creating trace2
trace2 =go.Scatter(
                    x = df2015.world_rank, #x eksenine 2015 ylndaki ilk 100 niversite koyulur.
                    y = df2015.total_score,  #y eksenine 2015 ylndaki alnt deerleri koyulur.
                    mode = "markers",      #marker modu seilir, yani nokta ekli
                    name = "2015",         #trace'in ad verilir.
                    marker = dict(color = 'rgba(255, 128, 2, 0.8)'),  #renk ve saydamlk verilir.
                    text= df2015.university_name)  #bir deerin zerine gelindiinde niversite ad gzkr.

# creating trace3
trace3 =go.Scatter(
                    x = df2016.world_rank, #x eksenine 2016 ylndaki ilk 100 niversite koyulur.
                    y = df2016.total_score,  #y eksenine 2016 ylndaki alnt deerleri koyulur.
                    mode = "markers",      #marker modu seilir, yani nokta ekli
                    name = "2016",         #trace'in ad verilir.
                    marker = dict(color = 'rgba(0, 255, 200, 0.8)'),  #renk ve saydamlk verilir.
                    text= df2016.university_name)  #bir deerin zerine gelindiinde niversite ad gzkr.

data = [trace1, trace2, trace3]  #trace1, trace2 ve trace3 data deikenine liste olarak atanr.

layout = dict(title = 'Total score vs world rank of top 100 universities with 2014, 2015 and 2016 years',
              xaxis= dict(title= 'World Rank',ticklen= 5,zeroline= False),
              yaxis= dict(title= 'Total Score',ticklen= 5,zeroline= False)  #balk, x ve y ekseninin bilgisi dictionary olarak layout deikenine atanr.
             )
fig = dict(data = data, layout = layout)  #data parametresine data deikeni, layout parametresine layout deikeni atanarak figr oluturulur.

iplot(fig) #figr plot edilir.
```
