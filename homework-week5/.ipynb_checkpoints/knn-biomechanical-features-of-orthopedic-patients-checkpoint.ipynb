```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("../input/biomechanical-features-of-orthopedic-patients/column_2C_weka.csv")
```

```python
data.head()
```

<a id = "2"></a><b>
# Data Visualization

```python
sns.heatmap(data.corr(),annot=True)
plt.show()
```

```python
sns.pairplot(data,hue="class")
plt.show()
```

```python
data.drop(['degree_spondylolisthesis'],axis=1,inplace=True)
```

```python
data['class'].value_counts()
```

```python
A = data[data['class'] == 'Abnormal']
N = data[data['class'] == 'Normal']
```

```python
data.info()
```

```python
plt.figure()

plt.scatter(A['pelvic_radius'],A['sacral_slope'],color='red',label='Abnormal',alpha=0.5)
plt.scatter(N['pelvic_radius'],N['sacral_slope'],color='blue',label='Normal',alpha=0.5)

plt.xlabel("pelvic_radius")
plt.ylabel("sacrol_slope")
plt.legend()
plt.show()
```

```python
data["class"] = [1 if each == "Normal" else 0 for each in data["class"]]
```

```python
y = data["class"]
x_data = data.drop(["class"], axis = 1)
```

<a id = "4"></a><b>
## Normalizatin

```python
x = (x_data - np.min(x_data))/(np.max(x_data) - np.min(x_data))
```

<a id = "5"></a><b>
## Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42) 
```

<a id = "6"></a><b>
# K-Nearest Neighbour (KNN)

```python
from sklearn.neighbors import KNeighborsClassifier
```

<a id = "7"></a><b>
## Find k value

```python
score_list = []

for each in range(1, 50):
    knn2 = KNeighborsClassifier(n_neighbors = each)
    knn2.fit(x_train, y_train)
    score_list.append(knn2.score(x_test, y_test))
    print("K =",each,"Accuracy:",knn2.score(x_test,y_test))

plt.plot(range(1, 50), score_list)
plt.xlabel("k value")
plt.ylabel("accurancy")
plt.show()
```

```python
knn = KNeighborsClassifier(n_neighbors = 15)
knn.fit(x_train, y_train)
prediction = knn.predict(x_test)

# score
print("{} knn accuracy score: {}".format(15, knn.score(x_test, y_test)))
```
